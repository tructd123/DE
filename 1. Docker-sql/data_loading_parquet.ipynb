```python
# create an open SQL database connection object or a SQLAlchemy connectable
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:7762117689@localhost:5432/ny_taxi")
engine.connect()
```

```python
# Generate CREATE SQL statement from scheme to validation
print(pd.io.sql.get_schema(df_taxi, name="yellow_taxi_data", con=engine))
```

```python
import pyarrow.parquet as pq

parquet_file = pq.ParquetFile("data/yellow_tripdata_2025-01.parquet")

# create batches of 100,000 for the paraquet file
batches_iter = parquet_file.iter_batches(batch_size=100000)
batches_iter
```

```python
df_iter = next(batches_iter).to_pandas()
df_iter
```

```python
df_iter.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace', index=False)
```
